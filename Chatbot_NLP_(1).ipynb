```
import spacy
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

nlp = spacy.load("en_core_web_md")
sia = SentimentIntensityAnalyzer()

memory = []   # 🔥 list to store last 3 inputs

print("Chatbot: Hello! Type 'exit' to stop")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    # COMMAND: Show NER from memory
    if user_input.lower() == "show ner":
        print("Chatbot: Showing NER from last inputs:")

        for sentence in memory:
            doc = nlp(sentence)
            print(f"Sentence: {sentence}")
            for ent in doc.ents:
                print(ent.text, "->", ent.label_)
        continue

    #  COMMAND: Show POS from memory
    if user_input.lower() == "show pos":
        print("Chatbot: Showing POS from last inputs:")

        for sentence in memory:
            doc = nlp(sentence)
            print(f"\nSentence: {sentence}")
            for token in doc:
                print(token.text, "->", token.pos_)
        continue

    #  Store input in memory
    memory.append(user_input)

    #  Keep only last 3 sentences
    if len(memory) > 3:
        memory.pop(0)

    #  NLP Processing
    doc = nlp(user_input)


    # Sentiment
    score = sia.polarity_scores(user_input)

    if score['compound'] > 0:
        print("Chatbot: Positive")
    elif score['compound'] < 0:
        print("Chatbot: Negative")
    else:
        print("Chatbot: Neutral")


    #  TEXT CLASSIFICATION (NEW)

    text = user_input.lower()

    if "how are you" in text:
        print("Chatbot: I am fine! How can I help you?")

    elif "hello" in text or "hi" in text:
        print("Chatbot: Hello! Nice to meet you!")

    elif "working" in text:
        print("Chatbot: I am chatbot based on simple core concepts of NLP.")

    elif "nlp" in text:
        print("Chatbot: NLP (Natural Language Processing) is a branch of AI that enables computers to understand, interpret, and generate human language.")

    elif "bye" in text:
        print("Chatbot: Goodbye! Have a nice day!")

    elif "weather" in text:
        print("Chatbot: Weather is nice today!")

    else:
        print("Chatbot: Tell me more!")
```
