```
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download required data (only first time)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

text = "Hello!!! My email is abc123@gmail.com. I am running late. Call me at 03121234567."

# Step 1: Extract Email using Regex
email = re.findall(r'\S+@\S+', text)

# Step 2: Extract Phone Number
phone = re.findall(r'\d{11}', text)

# Step 3: Remove special characters (clean text)
clean_text = re.sub(r'[^a-zA-Z\s]', '', text)

# Step 4: Tokenization
tokens = word_tokenize(clean_text)

# Step 5: Remove Stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word.lower() not in stop_words]

# Final Output
print("Original Text:", text)
print("Email:", email)
print("Phone:", phone)
print("Clean Text:", clean_text)
print("Tokens:", tokens)
print("Filtered Words:", filtered_words)
```
